```python
gauss_index = 6
inp_beams = np.load(data_dir + 'inp_beams.npy')
rec_beams = np.load(data_dir + 'res_beams.npy')
```

```python

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8))
ex = [-0.2, 0.2, -0.2, 0.2]
inp_abs = ax1.imshow(np.abs(inp_beams[gauss_index]), extent = ex)
inp_arg = ax3.imshow(np.angle(inp_beams[gauss_index]), cmap = 'hsv', interpolation= 'none', extent = ex)
res_abs = ax2.imshow(np.abs(rec_beams[gauss_index]), extent = ex)
res_arg = ax4.imshow(np.angle(rec_beams[gauss_index]), cmap = 'hsv', interpolation= 'none', extent = ex)

fig.colorbar(inp_abs, ax=ax1, fraction=0.046, pad=0.04)
fig.colorbar(inp_arg, ax=ax3, fraction=0.046, pad=0.04)
fig.colorbar(res_abs, ax=ax2, fraction=0.046, pad=0.04)
fig.colorbar(res_arg, ax=ax4, fraction=0.046, pad=0.04)
fig.tight_layout()
ax1.set_title('Transmitted')
ax2.set_title('Received')
st = fig.suptitle('Transmitted and Received Gaussian Beam', fontsize = 'x-large')
st.set_y(0.95)
fig.subplots_adjust(top=0.90)

ax1.locator_params(nbins=4)
ax2.locator_params(nbins=4)
ax3.locator_params(nbins=4)
ax4.locator_params(nbins=4)


ax1.set_ylabel('y(m)')
ax3.set_ylabel('y(m)')

ax3.set_xlabel('x(m)')
ax4.set_xlabel('x(m)')
plt.show()
```
